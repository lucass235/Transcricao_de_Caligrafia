```python
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
# Lê o arquivo de texto como um DataFrame
tabela = pd.read_csv("./iam_data/words.txt", sep=" ", header=None, names=["path", "value1", "value2", "value3", "value4", "value5", "value6", "value7", "labeling"], usecols=[0, 1, 2, 3, 4, 5, 6, 7, 8], index_col=False)
```

```python

# agora vou fazer um loop para salvar todas as imagens
for index, row in tabela.iterrows():
    
    # Lê o caminho da imagem
    loc = row["path"]
    part01 = loc.split("-")[0]
    part02 = part01 + "-" + loc.split("-")[1]
    
    # Lê a imagem
    image = cv2.imread("./iam_data/words/" + part01 + "/" + part02 + "/" + loc + ".png")
    plt.imshow(image)
    
    # Adiciona o nome acima do gráfico
    plt.text(20, -2, row["labeling"], color='red', fontsize=20)
    
    #salvar imagem
    plt.savefig(f'./imagens/{loc}.png')
    
    # Limpar o gráfico atual
    plt.clf();
    
    
```
